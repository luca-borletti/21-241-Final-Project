```julia
import LinearAlgebra

# HITS algorithm
# credit to Amy Langville and Carl D. Meyer's "Google's PageRank and Beyond"

w1 = ["w4", 1]
w2 = ["w1", "w4", 2]
w3 = ["w2", 3]
w4 = ["w3", "w5", 4]
w5 = ["w4", 5]
w6 = ["w7", 6]
w7 = ["w8", 7]
w8 = ["w6", 8]

baseSet = [w1, w2, w3, w4, w5, w6, w7, w8]

numNodes = size(baseSet)[1]

adjacencyM = [Float64(0) for i=1:numNodes, j=1:numNodes]

for page in baseSet
    wNumber = findfirst(x -> x == page, baseSet)
    for outLink in page[1:end-1]
        lNumber = parse(Int, outLink[end:end])
        adjacencyM[wNumber, lNumber] = Float64(1)
    end
end

# for col=1:numNodes             #TESTING CODE
#     print(adjacencyM[col, :], "\n")
# end

numIterations = 1000

hub = [Float64(1) for i=1:numNodes]
auth = [Float64(1) for i=1:numNodes] # sometimes people initialize hub and authority scores at 1's, 
# in "Google's PageRank and Beyond" it is initialized at 1/numNodes

for iteration=1:numIterations
    auth = LinearAlgebra.normalize(adjacencyM' * hub)
    hub = LinearAlgebra.normalize(adjacencyM * auth)
end

for col=1:numNodes
    print("authority: ", auth[col], " and hub: ", hub[col], "\n")
end

# ahScores = [Float64(1) for i=1:numNodes, j=1:2]    # initialize hub and authority vals

# k = 1000   # number of iterations

# for iteration=1:k
#     # authority score update
#     authNorm = Float64(0)
#     newAhScores = [Float64(0) for i=1:numNodes, j=1:2]
#     for page in baseSet
#         wNumber = findfirst(x -> x == page, baseSet)
#         authority = Float64(0)
#         for potentialLinker in baseSet
#             if page != potentialLinker
#                 for link in potentialLinker[1:end-1]
#                     linkNumber = parse(Int, link[end:end])
#                     if wNumber == linkNumber
#                         lNumber = findfirst(x -> x == potentialLinker, baseSet)
#                         authority += ahScores[lNumber, 2]
#                     end
#                 end
#             end
#         end
#         newAhScores[wNumber, 1] = authority
#         authNorm += authority^2
#     end

#     # for col=1:numNodes    #TESTING CODE
#     #     print(newAhScores[col, 1], "\n")
#     # end

#     # authority score normalization
#     authNorm = authNorm^(0.5)

#     for index=1:numNodes
#         newAhScores[index, 1] = newAhScores[index, 1] / authNorm
#     end

#     # for col=1:numNodes     #TESTING CODE
#     #     print(newAhScores[col, 1], "\n")
#     # end

#     # hub score update

#     hubNorm = Float64(0)

#     for page in baseSet
#         wNumber = findfirst(x -> x == page, baseSet)
#         hub = Float64(0)
#         for linkingTo in page[1:end-1]
#             linkNumber = parse(Int, linkingTo[end:end])
#             hub += ahScores[linkNumber, 1]
#         end
#         newAhScores[wNumber, 2] = hub
#         hubNorm += hub^2
#     end

#     # for col=1:numNodes     #TESTING CODE
#     #     print(newAhScores[col, 2], "\n")
#     # end

#     # hub score normalization
#     hubNorm = hubNorm^(0.5)

#     for index=1:numNodes
#         newAhScores[index, 2] = newAhScores[index, 2] / hubNorm
#     end

#     # for col=1:numNodes     #TESTING CODE
#     #     print(newAhScores[col, 2], "\n")
#     # end
    
#     ahScores = newAhScores
# end

# # for row=1:numNodes
# #     print(ahScores[row, :], "\n")
# # end

# for col=1:numNodes
#     nodeType = "a hub"
#     index = 2
#     if ahScores[col,:][1] > ahScores[col,:][2]
#         nodeType = "an authority"
#         index = 1
#     end
#     if round(ahScores[col,:][1], digits = 3) < 0.01
#         if round(ahScores[col,:][2], digits = 3) < 0.01
#             nodeType = "neither"
#         end
#     end
#     print("\n", "w", col, " is ", nodeType, " with score: ", round(ahScores[col,:][index], digits = 3)) 
# end
```
